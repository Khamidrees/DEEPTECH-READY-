```python
import torch
from transformers import CLIPProcessor, CLIPModel
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the pretrained CLIP model
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)
```

```python
from PIL import Image

# Example image URLs or paths
images = [
    Image.open("maggi original.jpeg"),
    Image.open("maggi fake.jpeg")
]

texts = ["Real Product", "Fake Product"]
```

```python
inputs = processor(
    text=texts,
    images=images,
    return_tensors="pt",
    padding=True
).to(device)

outputs = model(**inputs)

# Extract similarity scores
logits_per_image = outputs.logits_per_image
probs = logits_per_image.softmax(dim=1)

print(probs)
```

```python
texts = [
    "An authentic Maggi seasoning cube in official packaging",
    "A counterfeit or fake Maggi seasoning cube with incorrect packaging"
]

inputs = processor(
    text=texts,
    images=images,
    return_tensors="pt",
    padding=True
).to(device)

outputs = model(**inputs)
probs = outputs.logits_per_image.softmax(dim=1)

for idx, image in enumerate(images):
    image_probs = probs[idx]
    pred_label = texts[image_probs.argmax()]
    confidence = image_probs.max().item()
    print(f"Image {idx+1}: predicted as {pred_label} (confidence {confidence:.2f})")
```

```python
# Plot your test images and predictions
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
for i, img in enumerate(images):
    axes[i].imshow(img)
    axes[i].axis("off")
    axes[i].set_title(f"Predicted: {texts[probs[i].argmax()]}")
plt.show()
```

```python
# app.py

import streamlit as st
import torch
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import numpy as np

# Load CLIP model once
@st.cache_resource
def load_clip_model():
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(device)
    processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
    return model, processor, device

model, processor, device = load_clip_model()

st.title("Fake Product Detection with CLIP")

st.markdown("""
Upload an image of your product, and the model will predict whether it's **Real** or **Fake**.

 **How it works:**
- The model computes similarity between the uploaded image and the text prompts:
  - *Real Product*
  - *Fake Product*
- The label with the highest score is shown as the prediction.
""")

uploaded_file = st.file_uploader("Upload Product Image", type=["jpg", "jpeg", "png"])

if uploaded_file is not None:
    image = Image.open(uploaded_file).convert("RGB")
    st.image(image, caption="Uploaded Image", use_column_width=True)

    # Define your two text classes
    texts = ["Real Product", "Fake Product"]

    # Process inputs
    inputs = processor(
        text=texts,
        images=image,
        return_tensors="pt",
        padding=True
    ).to(device)

    # Get model outputs
    with torch.no_grad():
        outputs = model(**inputs)
        logits_per_image = outputs.logits_per_image
        probs = logits_per_image.softmax(dim=1).cpu().numpy().flatten()

    pred_idx = int(np.argmax(probs))
    pred_label = texts[pred_idx]
    confidence = probs[pred_idx]

    st.success(f"**Prediction:** {pred_label} (Confidence: {confidence:.2%})")

    # Show probability chart
    st.bar_chart({
        "Class": texts,
        "Probability": probs
    })
```
